```python
import platereadertools as prt
import numpy as np
```

```python
plate1, time1 = prt.Organize('../Data/Functional Screens/FS20230625ATc1c2.txt',
                      8,12,18,3/60)
plate2, time2 = prt.Organize('../Data/Functional Screens/FS20230625ATc3.txt',
                      8,12,18,3/60)
```

```python
reads = list(plate1.keys())
print(reads)
```

```python
time1.keys()
```

```python
time = dict()
for read in reads:
   time[read] = np.mean((time1[read], time2[read]), axis=0) 
```

```python
time
```

```python
for read in reads:
    prt.well_curves(plate1, time, read)
    prt.well_curves(plate2, time, read)
```
