```python
import platereadertools as prt
import numpy as np
```

```python
plate1, time1 = prt.Organize('../Data/Functional Screens/p2x11_MGz1_pATFXsgT6K_FS20230623a.txt',
                      8,12,18,3/60)
plate2, time2 = prt.Organize('../Data/Functional Screens/p2x11_MGz1_pATFXsgT6K_FS20230623b.txt',
                      8,12,18,3/60)
plate3, time3 = prt.Organize('../Data/Functional Screens/FS20230625R1C3.txt',8,12,18,3/60)
```

```python
reads = list(plate1.keys())
print(reads)
```

```python
time3.keys()
```

```python
time = dict()
for read in reads:
   time[read] = np.mean((time1[read], time2[read], time3[read]), axis=0) 
```

```python
time
```

```python
for read in reads:
    prt.well_curves(plate1, time, read)
    prt.well_curves(plate2, time, read)
    prt.well_curves(plate3, time, read)
```
